```python
#data
import pandas as pd
import numpy as np

#Plots
import matplotlib.pyplot as plt

#linear regiression nModel Building Library
import sklearn.linear_model
```

```python
data = pd.read_csv('yellow_tripdata_2020-01.csv', low_memory=False)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
df = pd.DataFrame(data[['PULocationID','DOLocationID']])
```

```python
df.info()
```

```python
X=df[['PULocationID']]
y= data.iloc[:, 1]
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
data['clusters'] = y_kmeans
```

```python
y_kmeans
```

```python
import datetime

x = pd.to_datetime(data['tpep_pickup_datetime'])
y = pd.to_datetime(data['tpep_dropoff_datetime'])
duration = y-x
data['duration']= duration.values.astype("float64")
```

```python
data
```

```python
total = pd.DataFrame(data[['trip_distance','duration','clusters']])
```

```python
total
```

```python
total['duration'] = pd.to_datetime(data['duration']).astype('int64')

X=total[['trip_distance','duration','clusters']]
y= data['fare_amount']
model= sklearn.linear_model.LinearRegression()
model.fit(X, y)
```

```python
y_predicted = model.predict(X) #use x to predict y
print(y_predicted)
```
